```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 
import os
```

Como já havia feito um trabalho sobre processamento em python optei por usar de novo e exercitar mais o opencv. Farei os dois exercícios nesse notebook usando opencv numpy e matplotlib como ferramentas principais. 

```python
im = cv2.imread('lena.jfif')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(im)
```

Vou usar numpy para fazer as contagens e matplotlib para gráficos

```python
im.shape
unique, counts = np.unique(im, return_counts=True)
val_count = pd.DataFrame({'val': unique, 'count': counts})
graph = sns.barplot(x='val', y='count',color='blue', data=val_count,label=None)
graph.set(xticks=[0,128,255])
```
