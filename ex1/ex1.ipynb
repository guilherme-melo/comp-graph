Vou usar numpy para fazer as contagens e matplotlib para gráficos

```python
im.shape
r,g,b = cv2.split(im)
unique, counts = np.unique(r, return_counts=True)
val_count = pd.DataFrame({'val': unique, 'count': counts})
graph = sns.barplot(x='val', y='count',color='blue', data=val_count,label=None)
graph.set(xticks=[0,128,255], title = "Red")


```

Convertendo pra HSL

```python
im = cv2.cvtColor(im, cv2.COLOR_RGB2HLS)
h,l,s = cv2.split(im)
```

```python
unique, counts = np.unique(l, return_counts=True)
val_count = pd.DataFrame({'val': unique, 'count': counts})
graph = sns.barplot(x='val', y='count',color='blue', data=val_count,label=None)
graph.set(xticks=[0,128,255],title="Luminance")
```

Equalizando o histograma.

```python
l = cv2.equalizeHist(l)

unique, counts = np.unique(l, return_counts=True)
val_count = pd.DataFrame({'val': unique, 'count': counts})
graph = sns.barplot(x='val', y='count',color='blue', data=val_count,label=None)
graph.set(xticks=[0,128,255],title="Luminance")
```

```python
im = cv2.merge((h,l,s))
im = cv2.cvtColor(im, cv2.COLOR_HLS2RGB)

plt.axis('off')
plt.imshow(im)
```
